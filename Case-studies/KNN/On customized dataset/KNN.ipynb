```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfx=pd.read_csv("xdata.csv")
X=dfx.values
X=X[:,1:]
print(X.shape)
```

```python
dfy=pd.read_csv("ydata.csv")
y=dfy.values
y=y[:,1:].reshape((-1,))
print(y.shape)

```

```python
query_x=np.array([2,3])
plt.scatter(X[:,0],X[:,1],c=y)  #plotting features of x
plt.scatter(query_x[0],query_x[1],color="red")    #plotting random point at(2,3)
plt.show()
```

```python
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))
```

```python
def knn(x,y,querypoint,k=5):
    vals=[]
    m=x.shape[0]
    for i in range(m):
        #print(x[i])
        d=dist(querypoint,x[i])
        vals.append((d,y[i]))         #storing distance and class as tuple in vals
    vals=sorted(vals)    #getting nearest point to the foreing point along with class of all nearest points
    vals=vals[:k]          #getting only k points
    vals=np.array(vals) 
    #converting to array
    #print(vals)
    new_vals=np.unique(vals[:,1],return_counts=True)    #it will only return distance and the frequency
    index=new_vals[1].argmax()                           #return class with maximum frequency
    pred=new_vals[0][index]
    #print(new_vals)
    return (pred)
```

```python
knn(X,y,query_x)
```
